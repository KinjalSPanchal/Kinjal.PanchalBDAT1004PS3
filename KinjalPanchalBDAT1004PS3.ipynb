```python
#1) 
import pandas as pd
import numpy as np
```

```python
#4)
users.groupby('occupation')['age'].mean()
```

```python
#5)
def male_ratio(series):
    male_count = (series == 'M').sum()
    total_count = len(series)
    return male_count / total_count
    
male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(male_ratio)
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)

print("male_ratio_per_occupation (sorted):/n", male_ratio_per_occupation)
```

```python
#6)
users.groupby('occupation')['age'].agg(['min', 'max'])
```

```python
#7)
users.groupby(['occupation', 'gender'])['age'].mean()
```

```python
#8)
def gender_percent(series):
    total_count = len(series)
    gender_counts = series.value_counts()
    percent = (gender_counts / total_count) * 100
    return percent

gender_percent_per_occupation = users.groupby('occupation')['gender'].apply(gender_percent)

print("\nPercentage of women and men per occupation:\n", gender_percent_per_occupation)
```

```python
#3)
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])
data.head()
```

```python
#9)
def stats(s):
    Min = s.min()
    Max = s.max()
    Mean = s.mean()
    Deviation = s.std()
    res = [Min, Max, Mean, Deviation]
    index = ["Min","Max","Mean","Deviation"]
    res_series = pd.Series(res,index=index)
    return res_series
loc_stats = data.apply(stats)
loc_stats
```

```python
#10)
def stats(row):
    row = row.dropna()
    numeric_values = pd.to_numeric(row, errors='coerce')  # Convert Timestamp to numeric, coerce errors to NaN
    return pd.Series({
        'Min': numeric_values.min(),
        'Max': numeric_values.max(),
        'Mean': numeric_values.mean(),
        'Deviation': numeric_values.std()
    })

day_stats = data.apply(stats, axis=1)
day_stats.head()

```

Que: 6 Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
us = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = us['Year']
B = us['Marriages_per_1000']
C = us['Divorces_per_1000']

plt.figure(figsize=(12,8))
plt.plot(A, B, label = 'Marriages per capita')
plt.plot(A, C, label = 'Divorces per capita')

plt.xlabel("Year", fontsize=20)
plt.ylabel("Rate (per capita)", fontsize=20)
plt.title("Marriage and divorces per capita in the U.S. between 1867 and 2014.", fontsize=20, fontweight='bold')
plt.legend(fontsize=15)
plt.grid(True)
plt.show()
```

Que: 7 Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!


```python
import pandas as pd
us = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us
```

```python
import numpy as np
import matplotlib.pyplot as plt

years = [1900, 1950, 2000]
marriages = us[us['Year'].isin(years)]['Marriages_per_1000']
divorces = us[us['Year'].isin(years)]['Divorces_per_1000']

plt.figure(figsize=(10, 6))
width = 0.35

plt.bar(np.arange(len(years)) - width/2, marriages, width, label='Marriages per capita')
plt.bar(np.arange(len(years)) + width/2, divorces, width, label='Divorces per capita')

plt.xlabel("Year", fontsize=14)
plt.ylabel("Rate (per capita)", fontsize=14)
plt.title("Comparison of Marriages and divorces per capita in the U.S. (1900, 1950, 2000)", fontsize=16, fontweight='bold')
plt.xticks(np.arange(len(years)),years)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
